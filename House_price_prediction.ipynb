```python
Location_stats=data.groupby('Location')['Location'].agg('count').sort_values(ascending=False)
```

```python
location_lessthan5 = Location_stats[Location_stats<=5]
```

```python
data['Location'] = data['Location'].apply(lambda x:'other' if x in location_lessthan5 else x)
```

```python
data['BHK'].unique()
```

```python
def extract_bedrooms(x):
    try:
        if isinstance(x, str):
            # If it's a full string like '3 BHK Apartment'
            if ' ' in x:
                return int(x.split(' ')[0])
            # If it's a string number like '3'
            return int(x)
        # If already a number
        elif isinstance(x, (int, float)):
            return int(x)
    except:
        return None  # return None if conversion fails (e.g., 'R')

data['bedrooms'] = data['BHK'].apply(extract_bedrooms)
```

```python
data
```

```python
data['Total sqft'].unique()
```

```python
data.describe()
```

```python
def clean_price(price):
    try:
        price = str(price).lower().strip()  # make lowercase and remove spaces
        
        if 'cr' in price:
            # Remove 'cr', convert to float, multiply by 1 crore (10 million)
            return float(price.replace('cr', '').strip()) * 1e7
        
        elif 'l' in price:
            # Remove 'l', convert to float, multiply by 1 lakh (100,000)
            return float(price.replace('l', '').strip()) * 1e5
        
        else:
            # Remove commas and convert directly to float
            return float(price.replace(',', ''))
    
    except:
        return None
```

```python
data['price'] = data['price'].apply(clean_price)
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['sqft_per_bedroom']=data['Total sqft']/data['bedrooms']
```

```python
data.sqft_per_bedroom.describe()
```

```python
data2 = data[data['sqft_per_bedroom'] >= 250]
```

```python
data2
```

```python
data2.price_sqft.describe()
```

```python
def clean_price_sqft(value):
    try:
        # Remove rupee symbol, commas, and whitespace
        value = str(value).replace('â‚¹', '').replace(',', '').strip()
        return float(value)
    except:
        return None
```

```python
data2['price_sqft'] = data2['price_sqft'].apply(clean_price_sqft)
```

```python
data2['price_sqft'].describe()
```

```python
data3 = data2[data2['price_sqft'] > 2000]
```

```python
data3
```

```python
data3.drop(columns=['BHK','sqft_per_bedroom','price_sqft'],axis=1,inplace=True)
```

```python
data3
```

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

```python
col_trans =make_column_transformer ((OneHotEncoder(sparse_output=False),['Location']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
scaler = StandardScaler()
```

```python
model = make_pipeline(col_trans, scaler,lr)
```

```python
data_input =data3.drop(columns = ['price'])
data_output=data3['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
input = pd.DataFrame([['Kurla',800.0,3.0]],columns = ['Location','Total sqft','bedrooms'])
```

```python
model.predict(input)
```
